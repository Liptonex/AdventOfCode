```python
with open('02_data.txt', 'r') as file:
    content = file.readlines()
content
```

```python
strategy = []
for line in content: 
    line = line.rstrip('\n')
    game = tuple(line.split(' '))
    strategy.append(game)
strategy
```

```python
def rock_paper_scisors(strategy = list):
    points = 0
    for game in strategy: 
        opponent = game[0]
        player = game[1]
        if player == 'X':
            points += 1
            if opponent == 'A':
                points += 3
            elif opponent == 'C':
                points += 6
        elif player == 'Y':
            points += 2
            if opponent == 'B':
                points += 3
            elif opponent == 'A':
                points += 6
        elif player == 'Z':
            points += 3
            if opponent == 'C':
                points += 3
            elif opponent == 'B':
                points += 6
    return points
```

```python
# (1 for Rock, 2 for Paper, and 3 for Scissors)
# plus the score for the outcome of the round (0 if you lost, 3 if the round was a draw, and 6 if you won).
```

```python
# The first column is what your opponent is going to play: A for Rock, B for Paper, and C for Scissors.
```

```python
# The second column, you reason, must be what you should play in response: 
# X for Rock, == 1
# Y for Paper == 2
# Z for Scissors == 3
```

```python
rock_paper_scisors(strategy)
```

```python
# "Anyway, the second column says how the round needs to end: 
# X means you need to lose, 
# Y means you need to end the round in a draw, 
# and Z means you need to win. Good luck!"
```

```python
def rock_paper_scisors_new(strategy = list):
    points = 0
    for game in strategy:
        opponent = game[0]
        goal = game[1]
        player = ''
        if opponent == 'A': #rock
            if goal == 'X': #lose
                player == 'Z'
                points += 3 + 0
            elif goal == 'Y': #draw
                player == 'X'
                points += 1 + 3
            elif goal == 'Z': #win
                player == 'Y'
                points += 2 + 6
        elif opponent == 'B': #paper
            if goal == 'X': #lose
                player == 'X'
                points += 1 + 0
            elif goal == 'Y': #draw
                player == 'Y'
                points += 2 + 3
            elif goal == 'Z': #win
                player == 'Z'
                points += 3 + 6
        elif opponent == 'C': #scisors
            if goal == 'X': #lose
                player == 'Y'
                points += 2 + 0
            elif goal == 'Y': #draw
                player == 'Z'
                points += 3 + 3
            elif goal == 'Z': #win
                player == 'X'
                points += 1 + 6
    return points
```
