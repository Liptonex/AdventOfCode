```python
import numpy as np
```

```python
with open('05_data.txt', 'r') as file:
    content = file.readlines()
content
```

```python
drawing_attempt = []
drawing = []
count = 0
for line in content: 
    if line == '\n':
        count += 1
        break
    else:
        count += 1
        drawing_attempt.append(line.rstrip('\n'))  
for n in range(1, len(drawing_attempt[0]), 4):
        drawing.append([])
for line in drawing_attempt:
    line = list(line)
    for i, n  in enumerate(range(1, len(line), 4)):
        if line[n] == ' ':
            continue
        else:
            drawing[i].append(line[n])
# print(drawing)
for n in range(0, len(drawing)):
    drawing[n].reverse()
    del drawing[n][0]
print(drawing)
```

```python
del content[0 : count]
```

```python
procedure = []
for line in content:
    line = line.rstrip('\n')
    line = line.split(' ')
    for item in line:
        try: 
            procedure.append(int(item))
        except ValueError:
            continue
procedure = np.array(procedure).reshape(-1,3)
procedure
```

```python
def crane(drawing, procedure):
    for line in procedure:
        num_of_boxes = line[0]*-1
        begin = line[1]-1
        end = line[2]-1
        crane = drawing[begin][num_of_boxes:]
        crane.reverse()
        print(crane)
        crane.reverse()
        for box in crane:
#             drawing[begin].reverse()
            drawing[begin].remove(box)
#             drawing[begin].reverse()
            drawing[end].append(box)
        print(drawing)
    return drawing
```

```python
end = crane(drawing, procedure)
```

```python
s = ''
for line in end:
    s += line[-1]
```
