```python
import string
```

```python
with open('03_data.txt', 'r') as file:
    content = file.readlines()
content
```

```python
rucksacks = []

for line in content: 
    line = line.rstrip('\n')
    rucksacks.append((set(line[:int(len(line)/2)]), set(line[int(len(line)/2):])))

rucksacks
```

```python
def repack_priority(rucksacks):
    summary = 0
    letters = list(string.ascii_lowercase + string.ascii_uppercase)
    for pack in rucksacks: 
        error = pack[0].intersection(pack[1])
        letter = error.pop()
        summary += letters.index(letter) + 1
#         print(letter, letters.index(letter) + 1, summary)
    return summary 
```

```python
repack_priority(rucksacks)
```

```python
def group_badges(rucksacks):
    summary = 0
    for i in range(0, len(rucksacks), 3):
        pack_1 = rucksacks[i][0].union(rucksacks[i][1])
        pack_2 = rucksacks[i+1][0].union(rucksacks[i+1][1])
        pack_3 = rucksacks[i+2][0].union(rucksacks[i+2][1])
        
        badge = pack_1.intersection(pack_2)
        badge = badge.intersection(pack_3)
        badge = badge.pop()
        letters = list(string.ascii_lowercase + string.ascii_uppercase)
        summary += letters.index(badge) + 1
    return summary
```

```python
group_badges(rucksacks)
```
